```python
# These are the "Tableau 20" colors as RGB.    
tableau20 = [(31, 119, 180), (174, 199, 232), (255, 127, 14), (255, 187, 120),    
             (44, 160, 44), (152, 223, 138), (214, 39, 40), (255, 152, 150),    
             (148, 103, 189), (197, 176, 213), (140, 86, 75), (196, 156, 148),    
             (227, 119, 194), (247, 182, 210), (127, 127, 127), (199, 199, 199),    
             (188, 189, 34), (219, 219, 141), (23, 190, 207), (158, 218, 229)]    
  
# Scale the RGB values to the [0, 1] range, which is the format matplotlib accepts.    
for i in range(len(tableau20)):    
    r, g, b = tableau20[i]    
    tableau20[i] = (r / 255., g / 255., b / 255.)   
    
plt.figure(figsize=(10, 10))    
  
# Remove the plot frame lines. They are unnecessary chartjunk.    
ax = plt.subplot(111)    
ax.spines["top"].set_visible(False)    
ax.spines["bottom"].set_visible(False)    
ax.spines["right"].set_visible(False)    
ax.spines["left"].set_visible(False) 


# Ensure that the axis ticks only show up on the bottom and left of the plot.    
# Ticks on the right and top of the plot are generally unnecessary chartjunk.    
ax.get_xaxis().tick_bottom()    
ax.get_yaxis().tick_left()    
  
# Limit the range of the plot to only where the data is.    
# Avoid unnecessary whitespace.    
plt.ylim(0.065, 0.145)    
plt.xlim(0, 1000)    
  
# Make sure your axis ticks are large enough to be easily read.    
# You don't want your viewers squinting to read your plot.    
plt.yticks(np.arange(0.065, 0.145, 0.005), [str(x*100) + "%" for x in np.arange(0.065, 0.145, 0.005)], fontsize=14)    
plt.xticks(fontsize=14)
plt.ylabel("Mean Square Error %", fontsize=16)
plt.xlabel("Iterations", fontsize=16)
# Provide tick lines across the plot to help your viewers trace along    
# the axis ticks. Make sure that the lines are light and small so they    
# don't obscure the primary data lines.    
for y in np.arange(0.065, 0.145, 0.005):    
    plt.plot(range(0, 1000), [y] * len(range(0, 1000)), "--", lw=0.5, color="black", alpha=0.3)    
  
  # Remove the tick marks; they are unnecessary with the tick lines we just plotted.    
plt.tick_params(axis="both", which="both", bottom="off", top="off",    
                labelbottom="on", left="off", right="off", labelleft="on") 
labels = ["Mean Square Error of Train Set", "Mean Square Error of Validation Set", "Mean Square Error of Test Set"]
data = ["MSE_trg", "MSE_val", "MSE_tst"]
for rank, column in enumerate(data):    
    # Plot each line separately with its own color, using the Tableau 20    
    # color set in order.    
    plt.plot(df_backprop["iteration"], df_backprop[column.replace("\n", " ")],lw=2.5, color=tableau20[rank], label = labels[rank])
    
plt.legend(loc = 7)
plt.text(500, 0.15, "Mean Square Error of NN with Backpropogation ", fontsize=17, ha="center")    
plt.show()
```
